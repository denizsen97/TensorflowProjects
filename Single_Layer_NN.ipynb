In this notebook, we are ploting the commonly used activation functions which are ReLU, Sigmoid, Tanh and Softmax.



```
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
```

Declare the input vector

```
input_vector = np.arange(-10, 10, 0.1)
```

Arange the ***ReLU*** visualization

```
#get the output of the function
relu_output = tf.nn.relu(input_vector)
#start the session
session = tf.Session()
plt.title("ReLU Activation Function")
plt.xlabel("Input values")
plt.ylabel("Output values")
plt.grid()
#plot the output
plt.plot(input_vector, session.run(relu_output))

session.close()
```

Arange the ***Tanh*** visualization

```
tanh_output = tf.nn.tanh(input_vector)
session = tf.Session()
plt.title("ReLU Activation Function")
plt.xlabel("Input values")
plt.ylabel("Output values")
plt.grid()
plt.plot(input_vector, session.run(tanh_output))
session.close()
```

Arange the ***Sigmoid*** visualization

```
sigmoid_output = tf.nn.sigmoid(input_vector)
session = tf.Session()
plt.title("ReLU Activation Function")
plt.xlabel("Input values")
plt.ylabel("Output values")
plt.grid()
plt.plot(input_vector, session.run(sigmoid_output))
session.close()
```

Arange the ***Softmax*** visualization

```
softmax_output = tf.nn.softmax(input_vector)
session = tf.Session()
plt.title("ReLU Activation Function")
plt.xlabel("Input values")
plt.ylabel("Output values")
plt.grid()
plt.plot(input_vector, session.run(softmax_output))
session.close()
```
